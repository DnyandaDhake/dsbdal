```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

# 1. Load and clean data
df = pd.read_csv(r'C:\Users\admin\Desktop\HousingData.csv')  # Adjust your file path if needed
df = df.fillna(df.median())  # Fill missing values with median

# 2. Prepare data
X = df.drop('MEDV', axis=1)  # Features (independent variables)
y = df['MEDV']  # Target prices (dependent variable)

# 3. Train model
model = LinearRegression()
model.fit(X, y)

# 4. Make predictions
df['Predicted'] = model.predict(X)

# 5. Calculate accuracy (RÂ² Score)
accuracy = r2_score(y, df['Predicted'])
print(f"Model Accuracy (RÂ² Score): {accuracy:.2%}")  # Shows as percentage

# 6. Calculate MSE and RMSE
mse = mean_squared_error(y, df['Predicted'])
rmse = np.sqrt(mse)
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")

# 7. Visualization 1: Scatter plot (Actual vs Predicted prices)
plt.figure(figsize=(10, 5))
plt.scatter(y, df['Predicted'], alpha=0.3, color='blue')
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--', linewidth=2)
plt.title('Actual Prices vs Predicted Prices')
plt.xlabel('Actual Price ($1000s)')
plt.ylabel('Predicted Price ($1000s)')
plt.grid(True)
plt.show()

# # 8. Visualization 2: Actual and Predicted prices in different colors
# plt.figure(figsize=(12, 6))
# plt.scatter(range(len(y)), y, color='red', label='Actual Price', alpha=0.7)
# plt.scatter(range(len(df['Predicted'])), df['Predicted'], color='blue', label='Predicted Price', alpha=0.7)

# plt.title('Actual vs Predicted Home Prices (Colored Comparison)')
# plt.xlabel('House Index')
# plt.ylabel('Price ($1000s)')
# plt.legend()
# plt.grid(True)
# plt.show()

# 9. Show sample predictions
print("\nSample Predictions (First 10 Houses):")
print(df[['MEDV', 'Predicted']].head(10))
```
